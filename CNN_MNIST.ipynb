```
# Fist cnn to detect hand writing
#Setup, Build, Load and Process Data, Compile, Test, Evaluate





#Setup
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



input_shape = (28,28,1)
output_classes = 10
loss_fxn = "categorical_crossentropy"
optimizer_fxn = tf.keras.optimizers.Adam()
metrics_fxn = ["accuracy"]
batch_size = 128
epochs = 5
verbose = 1
validation_split = 0.95



#Build
def build (input_shape, output_classes):
  model = tf.keras.Sequential()
  
  #First convolution layer, and if first layer, then use input shape
  model.add(tf.keras.layers.Conv2D(
      20, (5,5), strides = (2,2), activation = "relu", 
      input_shape = input_shape
  ))

  #add max pooling layer
  model.add(tf.keras.layers.MaxPooling2D(
          pool_size=(2, 2), strides=(2,2)
  ))

  model.add(tf.keras.layers.Dropout(0.2))
  
  #add another convolutional layer
  model.add(tf.keras.layers.Conv2D(
      50, (5,5), activation = "relu"
  ))

  #Add another max pooling layer
  model.add(tf.keras.layers.MaxPooling2D(
          pool_size=(2, 2), strides=(2,2)
  ))

  #flatten using relu
  model.add(tf.keras.layers.Flatten())

  #Densify
  model.add(tf.keras.layers.Dense(500, activation="relu"))
  #Bring it all together
  model.add(tf.keras.layers.Dense(output_classes, activation = "softmax"))
  return model


#Load and Process Data 
mnist_data = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist_data.load_data()
#print("Old x_train shape", x_train.shape)


#Reshape
x_train = x_train.reshape((60000,28,28,1))
x_test = x_test.reshape((10000,28,28,1))
# print("New x_train shape", x_train.shape)

#Normalize and Cast 
x_train = x_train.astype("float32")
x_train /= 255.0
x_test = x_test.astype("float32")
x_test /= 255.0


#To Categorical
y_train = tf.keras.utils.to_categorical(y_train, num_classes = output_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes = output_classes)


model = build(input_shape, output_classes)


#Compile
model.compile(optimizer= optimizer_fxn, loss = loss_fxn, metrics = metrics_fxn)

model.summary()


#Using TensorBoard
callbacks = [
             tf.keras.callbacks.TensorBoard(log_dir = './logs')
]





#Training
training = model.fit(x_train, y_train, batch_size = batch_size, 
                     epochs = epochs, verbose = verbose, 
                     callbacks = callbacks, validation_split=validation_split )



#Evaulating
score = model.evaluate(x_test, y_test, verbose = verbose)
print("/nTest Score: ", score[0])
print("Test Accuracy: ", score[1])
```
